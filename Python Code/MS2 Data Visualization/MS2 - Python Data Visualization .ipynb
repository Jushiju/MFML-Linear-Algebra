```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
demographics_df = pd.read_csv('Downloads/MFML/cleaned_demographic_data.csv')
transactions_df = pd.read_csv('Downloads/MFML/Cleaned_customer_transaction.csv')
social_interactions_df = pd.read_csv('Downloads/MFML/cleaned_Social_Interaction_data.csv')
```

```python
demographics_df['Age'] = pd.to_numeric(demographics_df['Age'], errors='coerce')
cleaned_age_data = demographics_df['Age'].loc[(demographics_df['Age'] > 0) & (demographics_df['Age'] <= 100)]
cleaned_age_data.describe()
```

```python
plt.figure(figsize=(12, 8))

# Plot the histogram with a KDE curve
sns.histplot(demographics_df['Age'], bins=30, kde=True, color='skyblue', edgecolor='black')

plt.title('Age Distribution in Demographics Dataset', fontsize=18)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.xticks(rotation=45, ha='right', fontsize=12)

plt.tight_layout()

plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

```python

data = {'Location': ['New York', 'Los Angeles', 'Chicago', 'New York', 'Los Angeles', 'Chicago', 'Miami', 'Miami', 'Dallas', 'Chicago']}
df = pd.DataFrame(data)

sns.set(style="whitegrid", palette="muted")

plt.figure(figsize=(12, 8)) 
ax = sns.countplot(x='Location', data=df, palette='coolwarm', linewidth=3, edgecolor='black')

plt.title('Location Distribution', fontsize=20, weight='bold', color='navy', pad=20)
plt.xlabel('Location', fontsize=16, weight='bold', color='darkred')
plt.ylabel('Count', fontsize=16, weight='bold', color='darkred')

plt.xticks(rotation=45, ha='right', fontsize=14, color='green')

plt.yticks(fontsize=14, color='purple')

plt.grid(True, axis='y', linestyle='--', color='gray', linewidth=0.5)


for p in ax.patches:
    p.set_edgecolor('black')
    p.set_linewidth(3)

# Add annotation to each bar to show the count on top of it
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2, height + 0.1, int(height), ha="center", fontsize=14, weight='bold', color='black')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(data=transactions_df, x='TransactionDate', y='Amount', color='green', linewidth=2)
plt.title('Transactions Over Time', fontsize=18)
plt.xlabel('Transaction Date', fontsize=14)
plt.ylabel('Transaction Amount', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(rotation=45, fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(data=transactions_df, x='PaymentMethod', palette='Set1')
plt.title('Transaction Count by Payment Method', fontsize=18)
plt.xlabel('Payment Method', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=demographics_df, x='IncomeLevel', palette='Set2', edgecolor='black', linewidth=1.5)
plt.title('Income Level Distribution', fontsize=22, weight='bold', color='darkblue')
plt.xlabel('Income Level', fontsize=16, weight='bold', color='darkblue')
plt.ylabel('Count', fontsize=16, weight='bold', color='darkblue')
plt.xticks(rotation=45, ha='right', fontsize=14, color='darkgreen')
plt.yticks(fontsize=14, color='darkgreen')
plt.grid(True, linestyle='--', alpha=0.7)
for p in plt.gca().patches:
    plt.gca().annotate(f'{int(p.get_height())}', 
                       (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha='center', va='center', 
                       fontsize=12, color='black', fontweight='bold', 
                       xytext=(0, 9), textcoords='offset points')
plt.tight_layout()
plt.show()
```

```python
g = sns.FacetGrid(cleaned_age_data, col='IncomeLevel', height=6, aspect=1.5, palette='Set2')
g.map(sns.histplot, 'Age', bins=30, kde=True, edgecolor='black')

g.set_axis_labels('Age', 'Frequency')
g.set_titles(col_template='{col_name} Income Level')

g.fig.suptitle('Age Distribution by Income Level', fontsize=20, weight='bold', color='darkblue')
g.fig.tight_layout()
g.fig.subplots_adjust(top=0.9)

plt.show()
```
